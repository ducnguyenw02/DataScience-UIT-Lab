# Câu 11 *Viết hàm trả về kết quả dự đoán từ một hoặc nhiều feature thuộc 'curb-weight', 'engine-size', 'length', 'width', 'horsepower', 'city-mpg', 'highwaympg', 'wheel-base', 'bore'

```python
# Tìm hiểu keyword arguments python
```

```python
def input_feature():
  print("Nhập số lượng feature: ") 
  a = int(input())
  list_feature = []
  for i in range(a):
    print("Feature thứ " + str(i + 1) + " :" )
    b = input()
    list_feature.append(b)
  return list_feature 
```

```python
def return_pre():
  feature = input_feature()
  X = df[feature]
  y = df['price']
 
  lr_feature = LinearRegression() 

  lr_feature.fit(X, y) 

  print("Kết quả mô hình hồi quy: ") 

  print("Kết quả intercept (b0): " + str(lr_feature.intercept_))
  print("Kết quả slope (b1)    : " + str(lr_feature.coef_))

  return lr_feature.predict(X)
```

```python
kq_pre = return_pre()

print(kq_pre)
```
