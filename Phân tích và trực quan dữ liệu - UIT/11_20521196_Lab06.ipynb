# Câu 1. Tạo dataframe (df) lưu trữ dataset tại đường dẫn trên. Xuất 5 samples đầu tiên. Dataset có bao nhiêu sample, số lượng bao nhiêu feature.

```python
df_excel = pd.ExcelFile("/content/drive/MyDrive/Data Scientist/DS105.N11 - Phân tích và trực quan dữ liệu/Canada.xlsx")
```

```python
df_excel.sheet_names
```

```python
df = df_excel.parse('Canada by Citizenship',skiprows = range(20), skipfooter=2)
df
```

# Câu 2. Đổi tên các column thành tên mới như sau: columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}. Sau đó loại bỏ các column ['AREA','REG','DEV','Type','Coverage'] ra khỏi dataset.


```python
df.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
```

```python
df = df.drop(columns = ['AREA','REG','DEV','Type','Coverage'])
```

#Câu 3. Tạo thêm cột 'Total' trong dataset, lưu trữ tổng số người từ năm 1980 đến 2013 của từng quốc gia.


```python
df['Total'] = df.sum(axis = 1)
df
```

# Câu 4. Chuyển đổi các giá trị trong cột Country thành index của dòng.

```python
df.set_index("Country", inplace = True)
df
```

# Câu 5 Gọi biến years lưu trữ thông tin các cột của các năm từ 1980 đến năm 2013 (years = [1980, 1981, …, 2013]). Thiết kế các cách truy vấn trả về một dataframe với input là một country bất kỳ.


```python
years = df.iloc[:, 3:37].columns

years 
```

```python
def df_country(country, y = years):
  for i in df.index:
    if i == country:
      print(df[years].loc[i])
```

```python
df_country(country = 'Japan', y = years)
```

# Câu 6. Tạo plot thể hiện số lượng di cư (sự biến động, diễn biến) của Hati và Japan đến Canada từ năm 1980 đến 2013

```python
import matplotlib.pyplot as plt

Hati  = df.iloc[df.index == 'Haiti'][years]
Japan = df.iloc[df.index == 'Japan'][years]
```

```python
plt.figure(figsize = (20,8))
plt.title('Số lượng di cư của Haiti và Japan đến Canada từ 1980 đến 2013')
plt.plot(Hati.columns,Hati.values.reshape(Hati.shape[1],), label = 'Haiti')
plt.plot(Japan.columns,Japan.values.reshape(Japan.shape[1],), label = 'Japan')
plt.show()
```

# Câu 7. Tạo plot mô tả xu hướng nhập cư của năm quốc gia có tỷ lệ nhập cư cao nhất vào Canada từ năm 1980 đến 2013.

```python
country_5 = df['Total'].sort_values(ascending=False)[0:5].index
data_top5 = df[years].transpose()[country_5]
```

```python
data_top5.plot(kind = 'area', stacked = False,figsize = (15,10))
plt.title("Top 5 quốc gia có lượng di cư lớn nhất vào Canada từ 1980 -> 2013")
plt.xlabel("Năm")
plt.ylabel("Lượng di cư")
#plt.plot(data_top5, kind = 'area')

plt.show()
```

# Câu 8. Tạo plot mô tả xu hướng nhập cư của năm quốc gia có tỷ lệ nhập cư thấp nhất vào Canada từ năm 1980 đến 2013.

```python
country_5 = df['Total'].sort_values(ascending=False)[-5:].index
data_top5 = df[years].transpose()[country_5]
```

```python
data_top5.plot(kind = 'area', stacked = False,figsize = (15,10))
plt.title("Top 5 quốc gia có lượng di cư thấp nhất vào Canada từ 1980 -> 2013")
plt.xlabel("Năm")
plt.ylabel("Lượng di cư")

plt.show()
```

# Câu 10. Tạo các plot thể hiện tỉ lệ nhập cư của mỗi lục địa vào Canada từ năm 1980 đến 2013.


```python
df_continent = df.groupby('Continent').sum()['Total']

df_continent.plot(kind = 'pie',figsize = (8,8),autopct = '%1.1f%%')
plt.title("Tỉ lệ nhập cư của mỗi lục địa vào Canada")
plt.axis('equal')
plt.show()
```
