# Câu 8


```python
import random 

# Điền khuyết cột normalized-losses 
df["normalized-losses"] = df["normalized-losses"].replace(to_replace = np.NaN, value= df['normalized-losses'].mean())

# Điền khuyết cột stroke 
df["stroke"] = df["stroke"].replace(to_replace = np.NaN, value= random.choice(np.array(df["stroke"])))

# Điền khuyết cột bore 
df["bore"] = df["bore"].replace(to_replace = np.NaN, value= df['bore'].mean())

# Điền khuyết cột horsepower 
df["horsepower"] = df["horsepower"].replace(to_replace = np.NaN, value= random.choice(np.array(df["horsepower"])))

# Điền khuyết cột peak-rpm 
df["peak-rpm"] = df["peak-rpm"].replace(to_replace = np.NaN, value= df['bore'].mean())

# Điền khuyết cột num-of-doors 
df["num-of-doors"] = df["num-of-doors"].replace(to_replace = np.NaN, value= pd.Series(df['num-of-doors'].value_counts()).sort_values(ascending=False).index[0])
```

```python
df.to_csv('df_cau08.csv')
```

# Câu 12

```python
# Chuẩn hóa "length", "width" và "height" trong phạm vi [0,1]

df['length'] = (df['length'] - df['length'].min()) / (df['length'].max() - df['length'].min())

df['width'] = (df['width'] - df['width'].min()) / (df['width'].max() - df['width'].min())

df['height'] = (df['height'] - df['height'].min()) / (df['height'].max() - df['height'].min())
```

```python
df.to_csv('df_cau12.csv')
```

# Câu 13

```python
# Tạo các bin là ['Low', 'Medium', 'High'] cho “horsepower” và “price”

bins = np.linspace(min(df['horsepower']-1 ), max(df['horsepower']),4)

label = ['low','medium','high']

df['horsepower-bin'] = pd.cut(df['horsepower'], bins, labels = label, include_lowest = True)

bins = np.linspace(min(df['price']-1 ), max(df['price']),4)

label = ['low','medium','high']

df['price-bin'] = pd.cut(df['price'], bins, labels = label, include_lowest = True)
```

```python
df.to_csv('df_cau13.csv')
```
